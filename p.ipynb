# Análise de Vendas de uma Concessionária com Pandas

## Objetivo
Analisar as vendas de uma concessionária, calcular comissões e identificar o faturamento de um vendedor específico.


```python
import pandas as pd

dataframe = pd.DataFrame() 
vendas_cf = pd.read_excel("vendas_concessionaria.xlsx")
display(vendas_cf)
```

```python
# Preenchendo a coluna "Comissao" com valores
vendas_cf['Comissao'] = vendas_cf['Preco_Venda'] * 0.01
display(vendas_cf)
```

```python
# Verificando valores vazios nas linhas das tabelas
vendas_cf[vendas_cf.isna().any(axis=1)]
# Adicionando a mensagem "Não encontrado" aos valores vazios
vendas_cf = vendas_cf.fillna("Não encontrado")
#display(vendas_cf)
# Puxando apenas as vendas do Carlos
vendas_carlos_cf = vendas_cf.loc[vendas_cf['Vendedor'] == 'Carlos', ['Vendedor', 'Preco_Venda', 'Comissao', 'Modelo_Veiculo', 'Tipo_Veiculo']]
display(vendas_carlos_cf)
# Conferindo o faturamento do Carlos
faturamento_carlos_cf = vendas_cf[vendas_cf["Vendedor"] == "Carlos"]["Preco_Venda"].sum()
display(faturamento_carlos_cf)                      
    



```
